### Change the CMENS product, variable, and region of study

```python
url = 'https://my.cmems-du.eu/thredds/dodsC/cmems_obs-sl_glo_phy-ssh_my_allsat-l4-duacs-0.25deg_P1D'
lon_slice = slice(-180,180)
lat_slice = slice(-60,-45)
time_slice = slice('2019-01-01', '2020-01-01')
var = ['adt']
```

```python
session = setup_session(username = usr, password = pas, check_url=url)
pydap_ds = open_url(url, session=session)
store = xr.backends.PydapDataStore(pydap_ds)
```

```python
xr.open_dataset(store)['adt'].sel(time=time_slice).sel(longitude=lon_slice, latitude=lat_slice)
```
